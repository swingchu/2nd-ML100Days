```python
# 設定 data_path, 並讀取 data
dir_data = './kaggle/app-store-apple-data-set'
f_apple = os.path.join(dir_data, 'AppleStore.csv')
apple = pd.read_csv(f_apple)
apple.head()
```

```python
# 找出 Games 類型的 data
app_game = apple[apple["prime_genre"] == "Games"]
print(app_game.shape)
app_game.head()
```
